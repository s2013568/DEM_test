```python
import matplotlib.pyplot as plt
import csv

# Function to read the CSV file and return the data
def read_csv_file(filename):
    agent_counts = []
    velocities = []
    densities = []
    
    with open(filename, mode='r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row
        for row in reader:
            agent_counts.append(int(row[0]))
            velocities.append(float(row[1]))
            densities.append(float(row[2]))
    
    return agent_counts, velocities, densities

# Function to plot velocity against density
def plot_velocity_vs_density(filename):
    agent_counts, velocities, densities = read_csv_file(filename)
    for i in range(len(agent_counts)):
        agent_counts[i] = (agent_counts[i] * 0.36) / 26
    
    
    # Create the plot
    plt.figure(figsize=(8, 6))
    plt.scatter(densities, velocities, color='blue', label="Velocity vs Density")
    # plt.plot(densities, velocities, color='blue', linestyle='--')
    
    # Add labels and title
    plt.title("Test Agent Velocity vs Density")
    plt.xlabel("agent area of control Density")
    plt.ylabel("Velocity")
    plt.grid(True)
    
    # Add legend
    plt.legend(loc="best")
    
    # Display the plot
    plt.show()

# Call the function to plot the graph
csv_filename = "Cannon Result.csv"
plot_velocity_vs_density(csv_filename)
```
