```python
import csv
import matplotlib.pyplot as plt

# Initialize lists to store data for each strategy
bottleneck_heights_1 = []
values_1 = []

bottleneck_heights_2 = []
values_2 = []

# Reading the first CSV file for Strategy 1
with open('Strat_1_data.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        bottleneck_heights_1.append(float(row['Bottleneck Heights']))
        values_1.append(float(row['Values']))

# Reading the second CSV file for Strategy 2
with open('Strat_2_data.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        bottleneck_heights_2.append(float(row['Bottleneck Heights']))
        values_2.append(float(row['Values']))

with open('Strat_2_data.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        bottleneck_heights_3.append(float(row['Bottleneck Heights']))
        values_3.append(float(row['Values']))
# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(bottleneck_heights_1, values_1, marker='o', label='Strategy 1')  # Plotting Strategy 1
plt.plot(bottleneck_heights_2, values_2, marker='o', label='Strategy 2')  # Plotting Strategy 2

# Adding title and labels
plt.title('Values vs. Bottleneck Heights')
plt.xlabel('Bottleneck Heights')
plt.ylabel('Values')
plt.grid()
plt.xticks(sorted(set(bottleneck_heights_1 + bottleneck_heights_2)))  # Set x-ticks to match heights
plt.legend()  # Show legend
plt.show()
```
