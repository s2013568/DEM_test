```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('C:\\Users\\Peter\\OneDrive - University of Edinburgh\Desktop\\velocity_std.csv', header=None, names=['Step', 'Std_Dev_Velocity', 'Other'])

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['Step'], df['Std_Dev_Velocity'], color='blue', marker='o', linestyle='-', linewidth=2, markersize=5)

# Add titles and labels
plt.title('Standard Deviation of Velocity vs Step, mu = 0.55', fontsize=16)
plt.xlabel('Step', fontsize=14)
plt.ylabel('Standard Deviation of Velocity', fontsize=14)

# Customize the grid and layout
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

# Show the plot
plt.show()
```
