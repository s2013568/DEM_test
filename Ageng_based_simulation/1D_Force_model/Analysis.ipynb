```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('C:\\Users\\Peter\\OneDrive - University of Edinburgh\Desktop\\velocity_std.csv', header=None, names=['Step', 'Std_Dev_Velocity', 'Other'])

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['Step'], df['Std_Dev_Velocity'], color='blue', marker='o', linestyle='-', linewidth=2, markersize=5)

# Add titles and labels
plt.title('Standard Deviation of Velocity vs Step, mu = 0.55', fontsize=16)
plt.xlabel('Step', fontsize=14)
plt.ylabel('Standard Deviation of Velocity', fontsize=14)

# Customize the grid and layout
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file into a DataFrame
df = pd.read_csv('C:\\Users\\Peter\\OneDrive - University of Edinburgh\\Desktop\\new_model_trajectory.csv', header=None, names=['Step', 'position', 'Other'])

# Calculate the differences between consecutive position values
position_diff = np.diff(df['position'])

# Set a threshold for detecting large jumps (indicative of wrap-around)
threshold = 100  # Adjust this value based on your data

# Start plotting
plt.figure(figsize=(10, 6))

# Loop through the data and plot segments where no large jumps occur
start_index = 0
for i in range(1, len(df)):
    if abs(position_diff[i-1]) > threshold:
        # Plot the segment between start_index and i
        plt.plot(df['position'][start_index:i], df['Step'][start_index:i], color='black', linestyle='-', linewidth=0.5)
        start_index = i  # Update the start_index for the next segment

# Plot the final segment
plt.plot(df['position'][start_index:], df['Step'][start_index:], color='black', linestyle='-', linewidth=0.1)

# Add titles and labels
plt.title('New Model trajectory, tau=1, init_sep = 1 (Rotated)', fontsize=16)
plt.xlabel('position trajectory', fontsize=14)
plt.ylabel('Step', fontsize=14)

plt.ylim(1e6, 2e6)  # Adjust y-axis limits

# Customize the grid and layout
# plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

# Show the plot
plt.show()
```
