```python
import matplotlib.pyplot as plt
from utils import *
# Testing and visualizing
N = 100  # Number of agents
Length = 10  # Length of the area
width = 5  # Height of the area

# Initialize positions and velocities
positions_velocities = init(N, Length, width)
x_coords = positions_velocities[0]
y_coords = positions_velocities[1]
dx_n = positions_velocities[2]
dy_n = positions_velocities[3]
print(positions_velocities)

# Plotting
plt.figure(figsize=(10, 5))
plt.scatter(x_coords, y_coords, color="blue", label="Agent Position")
plt.quiver(x_coords, y_coords, dx_n, dy_n, angles='xy', scale_units='xy', scale=0.1, color="red", alpha=0.6, label="Velocity")

plt.xlim(0, Length)
plt.ylim(0, width)
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.title("2D Agent Initialization Visualization")
plt.legend()
plt.grid()
plt.show()
```
